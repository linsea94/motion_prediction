```python
import pandas as pd
import numpy as np
import csv
import math
import matplotlib.pyplot as plt
import AngularGrid
```

```python
df = pd.read_csv('../split_data/003.csv')
columns = ['frame_id', 'agent_id', 'pos_x', 'pos_y', 'vel_x', 'vel_y', 'scene_id', 'label', 'timestamp']
```

```python
first_frame = df.iloc[0]['frame_id']
first_frame
```

```python
load_file = '../split_data/003.csv'
save_folder = '../split_data/'
file_name = load_file.split('/')[2]
file_name = file_name.split('.csv')[0]
```

```python
target = 6
FIRST = True
target_state=[]
target_output=[]
other_state=[]
first_frame = df.iloc[0][0]

for i in range(len(df)):
  ped_info = df.iloc[i]
  ped_id = ped_info[1]
  ped_frame = ped_info[0]
  if ped_id == target:

    if FIRST:
      FIRST = False
      target_state.append(ped_info)
    else:
      target_output.append(ped_info)

  elif ped_frame == first_frame:
    other_state.append(ped_info)
  else:
    pass

df_target_state = pd.DataFrame(target_state)
df_target_state.columns = columns

df_target_output = pd.DataFrame(target_output)
df_target_output.columns = columns
df_target_output.to_csv(save_folder+ 'output/' + str(file_name) + '_' + str(target) +'_OP.csv', header = True, index = False)

df_other_state = pd.DataFrame(other_state)
df_other_state.columns = columns
print("====================")
print("target_state")
print(df_target_state)
print("====================")
print("target_output")
print(df_target_output)
print("====================")
print("other_state")
print(df_other_state)
```

```python
target_pt = AngularGrid.get_pt(df_target_state.iloc[0])
ag = [AngularGrid.get_AngularGrid(df_other_state, target_pt, num_of_pieces=72, max_dist=8)]
ag = pd.DataFrame(ag)
ag.to_csv(save_folder+ 'output/' + str(file_name) + '_' + str(target) +'_AG.csv', header = True, index = False)
```
